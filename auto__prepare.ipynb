```python
import pandas as pd
import os
import numpy as np
from tqdm import tqdm
import PySimpleGUI as sg
```

```python
target= pd.read_csv('alfabattle2_train_target.csv')
```

```python
def get_pivot(df, name, agg_funcs = False):
    if not agg_funcs:
        agg_funcs = ['min','max', 'median', 'sum', 'var']
    return pd.pivot_table(df, values='amnt',
                index=['app_id'], columns=[name],
                aggfunc={'amnt': agg_funcs},
                 fill_value=0.0,
                margins_name = name)
```

```python
def group(data, func):
    df = data.groupby('app_id').agg(func)
    return df
```

```python
def re_index(feature, name):
    index = []
    for i in feature.columns:
        if i[0] != name:
            index.append('{} {} {}'.format(name, i[0], str(i[1])))
        else:
            index.append('{} {}'.format(name, str(i[1])))
    feature.columns = index
    return feature
```

```python
def parqet_preparing(link, first = True):
    df = pd.read_parquet(link, engine='pyarrow')
    features =  pd.DataFrame(data =  df['app_id'].unique(), columns = ['app_id'])
    #
    # amnt:
    # 
    
    name = 'amnt'
    agg_func_g = [ 'sum' , 'count', 'max']
    data = df[[name,'app_id']]
    feature = group(data, agg_func_g)

    feature = re_index(feature, name)     
    
    features = features.merge(feature, on = 'app_id')
    
    #
    # currency:
    #           
    name = 'currency'
    agg_func_g = [ 'sum', 'var', 'count']
    agg_func_p = ['median', 'sum', 'var', 'count', 'min', 'max']
               
    data = df[[name,'app_id']]
    
    feature = group(data, agg_func_g).merge(get_pivot(df, name, agg_funcs = agg_func_p), on = 'app_id')
        
    feature = re_index(feature, name)     
    
    features = features.merge(feature, on = 'app_id')
    
    #
    # operation_kind:
    #           
    name = 'operation_kind'
    agg_func_g = [ 'median', 'sum', 'var', 'count', 'min', 'max']
    agg_func_p = [ 'median', 'sum', 'var', 'count', 'min', 'max']
               
    data = df[[name,'app_id']]
    
    feature = group(data, agg_func_g).merge(get_pivot(df, name, agg_funcs = agg_func_p), on = 'app_id')
        
    feature = re_index(feature, name)     
    
    features = features.merge(feature, on = 'app_id')
    
    #          
    # card_type:
    #
    
    #
    # operation_type:
    #           
    name = 'operation_type'
    agg_func_g = [ 'median', 'sum', 'var', 'count', 'min', 'max']
    agg_func_p = ['median', 'sum', 'var', 'count', 'min', 'max']
               
    data = df[[name,'app_id']]
    
    feature = group(data, agg_func_g).merge(get_pivot(df, name, agg_funcs = agg_func_p), on = 'app_id')
        
    feature = re_index(feature, name)     
    
    features = features.merge(feature, on = 'app_id')
    
    #           
    # operation_type_group:
    #  
    
    name = 'operation_type_group'
    agg_func_g = ['median', 'sum', 'var', 'count', 'min', 'max']
    agg_func_p = ['median', 'sum', 'var', 'count', 'min', 'max']
               
    data = df[[name,'app_id']]
    
    feature = group(data, agg_func_g).merge(get_pivot(df, name, agg_funcs = agg_func_p), on = 'app_id')
        
    feature = re_index(feature, name)     
    
    features = features.merge(feature, on = 'app_id')
    
    #           
    # ecommerce_flag:
    # 
    
    name = 'ecommerce_flag'
    agg_func_g = [ 'median', 'sum', 'var', 'count', 'min', 'max']
    agg_func_p = [ 'median', 'sum', 'var', 'count', 'min', 'max']
               
    data = df[[name,'app_id']]
    
    feature = group(data, agg_func_g).merge(get_pivot(df, name, agg_funcs = agg_func_p), on = 'app_id')
        
    feature = re_index(feature, name)     
    
    features = features.merge(feature, on = 'app_id')
    
    #           
    # payment_system:
    #
    
    name = 'payment_system'
    agg_func_g = [ 'median', 'sum', 'var', 'count', 'min', 'max']
    agg_func_p = ['median', 'sum', 'var', 'count', 'min', 'max']
               
    data = df[[name,'app_id']]
    
    feature = group(data, agg_func_g).merge(get_pivot(df, name, agg_funcs = agg_func_p), on = 'app_id')
        
    feature = re_index(feature, name)     
    
    features = features.merge(feature, on = 'app_id')
    
    #           
    # income_flag:
    # 
    
    name = 'income_flag'
    agg_func_g = ['median', 'sum', 'var', 'count', 'min', 'max']
    agg_func_p = ['median', 'sum', 'var', 'count', 'min', 'max']
               
    data = df[[name,'app_id']]
    
    feature = group(data, agg_func_g).merge(get_pivot(df, name, agg_funcs = agg_func_p), on = 'app_id')
        
    feature = re_index(feature, name)     
    
    features = features.merge(feature, on = 'app_id')
    
    #           
    # mcc:
    #  
    
    name = 'mcc'
    agg_func_g = ['median', 'sum', 'var', 'count', 'min', 'max']
    agg_func_p = ['median', 'sum', 'var', 'count', 'min', 'max']
               
    data = df[[name,'app_id']]
    
    feature = group(data, agg_func_g).merge(get_pivot(df, name, agg_funcs = agg_func_p), on = 'app_id')
        
    feature = re_index(feature, name)     
    
    features = features.merge(feature, on = 'app_id')
    
    #           
    # country:
    # 
    
    name = 'country'
    agg_func_g = [ 'median', 'sum', 'var', 'count', 'min', 'max']
    agg_func_p = ['median', 'sum', 'var', 'count', 'min', 'max']
               
    data = df[[name,'app_id']]
    
    feature = group(data, agg_func_g).merge(get_pivot(df, name, agg_funcs = agg_func_p), on = 'app_id')
        
    feature = re_index(feature, name)     
    
    features = features.merge(feature, on = 'app_id')
    
    #           
    # city:
    #
    
    #
    # mcc_category:
    #   
    
    name = 'mcc_category'
    agg_func_g = [ 'median', 'sum', 'var', 'count', 'min', 'max']
    agg_func_p = ['median', 'sum', 'var', 'count', 'min', 'max']
               
    data = df[[name,'app_id']]
    
    feature = group(data, agg_func_g).merge(get_pivot(df, name, agg_funcs = agg_func_p), on = 'app_id')
        
    feature = re_index(feature, name)     
    
    features = features.merge(feature, on = 'app_id')
    
    #           
    # days_before:
    #
    
    name = 'days_before'
    agg_func_g = ['median', 'sum', 'var', 'count', 'min', 'max']
               
    data = df[[name,'app_id']]
    
    feature = group(data, agg_func_g)
            
    feature = re_index(feature, name)  
    
    feature['days_before diff'] = feature['days_before max'] - feature['days_before min']
    
    features = features.merge(feature, on = 'app_id')
    
    # 
    # weekofyear:
    #
    
    #
    # hour_diff:
    #    
    
    name = 'hour_diff'
    agg_func_g = ['median', 'sum', 'var', 'count', 'min', 'max']
               
    data = df[[name,'app_id']]
    
    feature = group(data, agg_func_g)
        
    feature = re_index(feature, name)    
    
    feature['hour_diff diff'] = feature['hour_diff max'] - feature['hour_diff min']
    
    features = features.merge(feature, on = 'app_id')
    
    #
    # transaction_number:
    # 
    
    name = 'transaction_number'
    agg_func_g = ['median', 'sum', 'var', 'count', 'min', 'max']
               
    data = df[[name,'app_id']]
    
    feature = group(data, agg_func_g)
            
    feature = re_index(feature, name)     
    
    features = features.merge(feature, on = 'app_id')
    
    return features
```

```python
size = 2
input_folder = 'train'
output_folder = 'features/data'
files = os.listdir(input_folder)
for i in tqdm(range(0, size)):
    link = files[i]
    data = parqet_preparing('train/{}'.format(link)).merge(target, on = 'app_id')
    data.to_csv('{}/data_{}.csv'.format(output_folder, i))
    print(data.shape)
```
